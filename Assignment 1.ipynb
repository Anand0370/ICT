```python
#importing the required packages
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the dataset to python environment

```python
#loading the given excel data 
data = pd.read_excel('iris.xls')
data
```

# Display the columns in the dataset.

```python
data.columns
```

# Calculate the mean of each column of the dataset.

```python
#creating dataframe using the given data
df = pd.DataFrame(data)
df.mean(axis=0,numeric_only=True)
```

# Check for the null values present in the dataset

```python
df.info()
```

```python
df.isnull().any(axis=0)
#df.isnull().sum()
```

# Perform meaningful visualizations using the dataset. 

# Countplot

```python
#setting the required plot style
plt.style.use('seaborn')
sns.countplot(x ='Classification', data = df )
#adding labels for x and y axis
plt.xlabel('Classification', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
#adding tiltle for graph
plt.title('Iris Data Count', fontsize = 20)
#setting the required limits for x and y axis
plt.ylim(bottom = 0, top = 100)
#adjusting x and y ticks size
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show();
```

# Boxplot

```python
#setting the required plot style
plt.style.use('bmh')
#Defining a function graph()
def graph(y):
    sns.boxplot(x = 'Classification', y = y , data = df)
    plt.xlabel('Classification', fontsize = 18)
    plt.ylabel( y, fontsize = 18)
plt.figure(figsize=(20,20))
#plotting subplots using the function graph()
plt.subplot(221)
graph('SL')
plt.subplot(222)
graph('SW')
plt.subplot(223)
graph('PL')
plt.subplot(224)
graph('PW')
plt.show();
```

# Violinplot

```python
#setting the required plot style
plt.style.use('ggplot')
#Defining a function graph()
def graph(y):
    sns.violinplot(x = 'Classification', y = y , data = df)
    plt.xlabel('Classification', fontsize = 15)
    plt.ylabel( y, fontsize = 15)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
plt.figure(figsize=(15,15))
#plotting subplots using the function graph()
plt.subplot(221)
graph('SL')
plt.subplot(222)
graph('SW')
plt.subplot(223)
graph('PL')
plt.subplot(224)
graph('PW')
plt.show();
```

# Scatterplot

```python
#setting the required plot style
plt.style.use('classic')
def graph(x,y):
    sns.scatterplot(x = x, y = y , hue= 'Classification' , data = df, sizes=(40,40))
    plt.title((x,'vs',y), fontsize = 20)
    plt.xlabel( x, fontsize = 15)
    plt.ylabel( y, fontsize = 15)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
plt.figure(figsize=(20,20))
plt.subplot(221)
graph('SL','SW')
plt.subplot(222)
graph('PL','PW')
plt.show();
```

# Pairplot

```python
#setting the required plot style
plt.style.use('seaborn')
sns.pairplot(hue='Classification', data=df);
plt.show()
```
